```python
#Convert a series of date-strings to a timeseries
import pandas as pd


dates = pd.Series(['2026-01-10', '2026-02-15', '2026-03-20'])


timeseries = pd.to_datetime(dates)

print(timeseries)
```

```python
#DataFrame Merging and Joining
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aman', 'Riya', 'Karan', 'Simran']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4, 5],
    'Marks': [85, 90, 88, 76]
})

print("DF1")
print(df1)

print("DF2")
print(df2)
```

```python
inner_merge = pd.merge(df1, df2, on='ID', how='inner')

print(inner_merge)
```

```python
left_join = pd.merge(df1, df2, on='ID', how='left')

print(left_join)
```

```python
right_join = pd.merge(df1, df2, on='ID', how='right')

print(right_join)
```

```python
df1_index = df1.set_index('ID')
df2_index = df2.set_index('ID')

joined_df = df1_index.join(df2_index)

print(joined_df)
```

```python
#Merging with Multiple Keys
df3 = pd.DataFrame({
    'ID': [1, 1, 2, 2],
    'Section': ['A', 'B', 'A', 'B'],
    'Marks': [80, 85, 90, 95]
})

df4 = pd.DataFrame({
    'ID': [1, 2, 2],
    'Section': ['A', 'A', 'B'],
    'Grade': ['B', 'A', 'A+']
})

multi_merge = pd.merge(df3, df4, on=['ID', 'Section'])

print(multi_merge)
```

```python
#Concatenation and Merge
df_a = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Riya']
})

df_b = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Karan', 'Simran']
})

df_c = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'City': ['Jaipur', 'Delhi', 'Mumbai', 'Pune']
})
```

```python
concat_df = pd.concat([df_a, df_b], axis=0)

print(concat_df)
```

```python
final_merge = pd.merge(concat_df, df_c, on='ID')

print(final_merge)
```
