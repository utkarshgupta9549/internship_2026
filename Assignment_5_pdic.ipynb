```python
# 1) Repeat a tuple three times using * operator

t = (1, 2, 3)
result = t * 3

print("Original Tuple:", t)
print("Repeated Tuple:", result)
```

```python
# 2) Join three separate tuples into one new tuple using + operator

t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

new_tuple = t1 + t2 + t3

print("Joined Tuple:", new_tuple)
```

```python
# 3) Check whether a specific element exists inside a tuple

numbers = (10, 20, 30, 40, 50)
element = 30

if element in numbers:
    print(element, "exists in the tuple")
else:
    print(element, "does not exist in the tuple")
```

```python
# 4) Find total, highest and lowest value without sum(), max(), min()

nums = (12, 45, 8, 23, 67, 3)

total = 0
highest = nums[0]
lowest = nums[0]

for n in nums:
    total += n

    if n > highest:
        highest = n

    if n < lowest:
        lowest = n

print("Tuple:", nums)
print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

```python
# 5) Filter tuple and keep only values greater than 10

n = (3, 14, 7, 22, 9, 41, 18, 5)

filtered = ()

for value in n:
    if value > 10:
        filtered += (value,)

print("Original Tuple:", n)
print("Filtered Tuple:", filtered)
```

```python
# 6) Count elements in a set without using len()

s = {"cat", "dog", "bird", "fish"}

count = 0

for item in s:
    count += 1

print("Set:", s)
print("Number of elements:", count)
```

```python
# 7) Combine two sets into one set

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

combined_set = set1.union(set2)

print("Combined Set:", combined_set)
```

```python
# 8) Find common elements in two sets

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

common = s1.intersection(s2)

print("Common Elements:", common)
```

```python
# 9) Find elements present in either set but not both

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

result = s1.symmetric_difference(s2)

print("Elements in either set but not both:", result)
```
