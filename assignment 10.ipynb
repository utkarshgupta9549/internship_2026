```python
import numpy as np

# Create array
arr = np.array([[6, -8, 73, -110],
                [np.nan, -8, 0, 94]])

print("Original Array:")
print(arr)

# Replace NaN with 0
arr = np.nan_to_num(arr, nan=0)

print("\nAfter Replacing NaN with 0:")
print(arr)

# Interchange rows and columns (Transpose)
transposed = arr.T

print("\nAfter Interchanging Rows and Columns:")
print(transposed)
```

```python
import numpy as np

arr = np.arange(24).reshape(2, 3, 4)

print("Original Shape:", arr.shape)

# Move axis 0 to position 2
new_arr = np.moveaxis(arr, 0, 2)

print("New Shape:", new_arr.shape)

print("\nNew Array:")
print(new_arr)
```

```python
import numpy as np

arr = np.array([[10, 20, np.nan],
                [30, np.nan, 50],
                [40, 60, 70]])

print("Original Array:")
print(arr)

# Column means ignoring NaN
col_mean = np.nanmean(arr, axis=0)

# Find NaN positions
inds = np.where(np.isnan(arr))

# Replace NaN with column means
arr[inds] = np.take(col_mean, inds[1])

print("\nArray After Replacing NaN with Column Mean:")
print(arr)
```

```python
import numpy as np

arr = np.array([10, -5, 20, -15, 30])

print("Original Array:")
print(arr)

arr = np.where(arr < 0, 0, arr)

print("\nAfter Replacing Negative Values with 0:")
print(arr)
```

```python
import numpy as np

arr = np.array([10, -5, 20, -15, 30])

arr[arr < 0] = 0

print(arr)
```
