```python
import numpy as np
```

```python
# Create NumPy array
arr = np.array([10, 20, 30, 40, 50, 60])
print("Original 1D Array:")
print(arr)

# Convert 1D array to 2D
arr2d = arr.reshape(2, 3)
print("\n1D to 2D Array:")
print(arr2d)

# Print Array Attributes
print("\nArray Attributes:")
print("Shape:", arr2d.shape)
print("Dimension:", arr2d.ndim)
print("Data Type:", arr2d.dtype)
print("Item Size:", arr2d.itemsize)
```

```python
# Create a 3×3 NumPy array of all 9
arr_9 = np.full((3, 3), 9)
print("\n3x3 Array of 9:")
print(arr_9)
```

```python
#  Create a 1D array of 10 evenly spaced values between 25 and 125
even_space = np.linspace(25, 125, 10)
print("\n10 Evenly Spaced Values:")
print(even_space)
```

```python
# Convert Python List into NumPy Array
py_list = [5, 10, 15, 20]
np_arr = np.array(py_list)
print("\nPython List to NumPy Array:")
print(np_arr)
```

```python
# Reverse a 1D NumPy Array
reverse_arr = np_arr[::-1]
print("\nReversed Array:")
print(reverse_arr)
```

```python
#  Create a 4×4×3 array and extract value
arr3d = np.arange(48).reshape(4, 4, 3)
print("\n4x4x3 Array:")
print(arr3d)

value = arr3d[1, 0, -1]
print("\nSecond Set, First Row, Last Column:")
print(value)
```

```python
#  Create a 4×4 array and extract Odd Rows and Even Columns
arr4x4 = np.arange(1, 17).reshape(4, 4)
print("\n4x4 Array:")
print(arr4x4)

odd_rows_even_cols = arr4x4[::2, 1::2]
print("\nOdd Rows and Even Columns:")
print(odd_rows_even_cols)
```

```python
#  Slice first two rows and first two columns of second set
slice_arr = arr3d[1, :2, :2]
print("\nFirst Two Rows and First Two Columns of Second Set:")
print(slice_arr)
```

```python
#  Replace all odd numbers with -1 using for loop
arr_replace = np.array([[23, 56, 78, 93],
                        [71, 82, 13, 24]])

for i in range(arr_replace.shape[0]):
    for j in range(arr_replace.shape[1]):
        if arr_replace[i, j] % 2 != 0:
            arr_replace[i, j] = -1

print("\nReplace Odd Numbers with -1:")
print(arr_replace)
```

```python
#  Get indices of non-zero elements
arr_nonzero = np.array([1, 0, 2, 0, 3, 0, 4])
indices = np.nonzero(arr_nonzero)
print("\nIndices of Non-Zero Elements:")
print(indices)
```

```python
#  Arithmetic Operations
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("\nAddition:")
print(a + b)

print("\nMultiplication:")
print(a * b)
```

```python
#  Dot Product
arr1 = np.array([15, 20, 25])
arr2 = np.array([10, 40, 37])

dot_product = np.dot(arr1, arr2)

print("\nDot Product:")
print(dot_product)
```
