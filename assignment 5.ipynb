```python
import csv

# Create CSV file
file = open("address_book.csv", "w", newline='')

writer = csv.writer(file)

# Writing column headings
writer.writerow(["Name", "Address", "Mobile", "Email"])

# Taking input from user
n = int(input("How many records do you want to enter? "))

for i in range(n):
    print(f"\nEnter details for Record {i+1}")

    name = input("Enter Name: ")
    address = input("Enter Address: ")
    mobile = input("Enter Mobile Number: ")
    email = input("Enter Email: ")

    writer.writerow([name, address, mobile, email])

print("\nData inserted successfully into address_book.csv")

file.close()
```

```python
import sqlite3

# Connect database
conn = sqlite3.connect("college.db")

# Create cursor object
cur = conn.cursor()

# Create Student Table
cur.execute("""
CREATE TABLE IF NOT EXISTS student(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    city TEXT,
    marks INTEGER
)
""")

# Create Course Table
cur.execute("""
CREATE TABLE IF NOT EXISTS course(
    cid INTEGER PRIMARY KEY AUTOINCREMENT,
    course_name TEXT,
    duration TEXT
)
""")

# Create Teacher Table
cur.execute("""
CREATE TABLE IF NOT EXISTS teacher(
    tid INTEGER PRIMARY KEY AUTOINCREMENT,
    teacher_name TEXT,
    subject TEXT
)
""")

print("Tables created successfully")

conn.commit()
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("college.db")
cur = conn.cursor()

# Insert into student table
cur.execute("INSERT INTO student(name, city, marks) VALUES('Rahul', 'Jaipur', 85)")
cur.execute("INSERT INTO student(name, city, marks) VALUES('Aman', 'Delhi', 90)")
cur.execute("INSERT INTO student(name, city, marks) VALUES('Priya', 'Mumbai', 88)")

# Insert into course table
cur.execute("INSERT INTO course(course_name, duration) VALUES('Python', '3 Months')")
cur.execute("INSERT INTO course(course_name, duration) VALUES('Java', '6 Months')")

# Insert into teacher table
cur.execute("INSERT INTO teacher(teacher_name, subject) VALUES('Sharma Sir', 'Python')")
cur.execute("INSERT INTO teacher(teacher_name, subject) VALUES('Verma Sir', 'Java')")

conn.commit()

print("Records inserted successfully")

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("college.db")
cur = conn.cursor()

# Select all records
print("All Students")
data = cur.execute("SELECT * FROM student")

for row in data:
    print(row)

# Select specific columns
print("\nStudent Names and Marks")
data = cur.execute("SELECT name, marks FROM student")

for row in data:
    print(row)

# Select with condition
print("\nStudents with marks greater than 85")
data = cur.execute("SELECT * FROM student WHERE marks > 85")

for row in data:
    print(row)

conn.close()
```
