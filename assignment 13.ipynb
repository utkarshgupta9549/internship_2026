```python
import numpy as np

arr = np.array([[6, -8, 73, -110],
                [np.nan, -8, 0, 94]])

print("Original Array:")
print(arr)

# Replace NaN with 0
arr = np.nan_to_num(arr, nan=0)

print("\nAfter Replacing NaN with 0:")
print(arr)

# Interchange rows and columns (Transpose)
transposed = arr.T

print("\nAfter Interchanging Rows and Columns:")
print(transposed)
```

```python
import numpy as np

arr = np.array([[10, np.nan, 30],
                [20, 25, np.nan],
                [30, 35, 50]])

print("Original Array:")
print(arr)

col_mean = np.nanmean(arr, axis=0)

inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_mean, inds[1])

print("\nArray after replacing NaN with column average:")
print(arr)
```

```python
import numpy as np

arr = np.array([10, -5, 20, -8, 30])

print("Original Array:")
print(arr)

arr[arr < 0] = 0

print("\nAfter Replacing Negative Values:")
print(arr)
```

```python
import numpy as np
from scipy import stats

arr1 = np.array([[3, 4],
                 [5, 6]])

arr2 = np.array([[1, 0],
                 [7, 8]])

avg = (arr1 + arr2) / 2

print("Average Array:")
print(avg)

combined = np.concatenate((arr1.flatten(), arr2.flatten()))

print("\nMean =", np.mean(combined))
print("Median =", np.median(combined))
print("Mode =", stats.mode(combined, keepdims=True)[0][0])
```

```python
import numpy as np

A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])

B = np.array([9, -6, 17])

solution = np.linalg.solve(A, B)

print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])
```

```python
import numpy as np

A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])

B = np.array([[9],
              [-6],
              [17]])

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print("Solution:")
print(X)
```
