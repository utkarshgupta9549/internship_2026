```python
import pandas as pd

data = {
    "Name": ["Rahul", "Aman", "Priya"],
    "Marks": [85, 90, 95]
}

df = pd.DataFrame(data)

for index, row in df.iterrows():
    print(index, row["Name"], row["Marks"])
```

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Aman", "Priya"],
    "Marks": [85, 90, 95]
}

df = pd.DataFrame(data)

for row in df.itertuples():
    print(row.Name, row.Marks)
```

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Aman", "Priya"],
    "Marks": [85, 90, 95]
}

df = pd.DataFrame(data)

result = df[df["Marks"] > 85]

print(result)
```

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Aman", "Priya", "Rohit"],
    "Marks": [85, 90, 95, 80]
}

df = pd.DataFrame(data)

# Drop rows where Marks < 90
df = df[df["Marks"] >= 90]

print(df)
```
