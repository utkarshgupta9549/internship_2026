```python
import requests

city = input("Enter city name: ")

api_key = "YOUR_API_KEY"

url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=e905dfae640533d9d537ffbda87f7f93&units=metric"

response = requests.get(url)

data = response.json()

if data["cod"] == 200:

    print("\n------ Weather Report ------")

    print("City:", data["name"])
    print("Country:", data["sys"]["country"])

    print("Temperature:", data["main"]["temp"], "°C")
    print("Feels Like:", data["main"]["feels_like"], "°C")

    print("Minimum Temp:", data["main"]["temp_min"], "°C")
    print("Maximum Temp:", data["main"]["temp_max"], "°C")

    print("Humidity:", data["main"]["humidity"], "%")
    print("Pressure:", data["main"]["pressure"], "hPa")

    print("Weather:", data["weather"][0]["description"])

    print("Wind Speed:", data["wind"]["speed"], "m/s")
    print("Wind Direction:", data["wind"]["deg"], "°")

    print("Visibility:", data["visibility"], "meters")

    print("Sunrise Time:", data["sys"]["sunrise"])
    print("Sunset Time:", data["sys"]["sunset"])

else:
    print("City not found!")
```

```python
import random

choices = ["rock", "paper", "scissors"]

count_comp = 0
count_player = 0

while count_comp < 3 and count_player < 3:
    user_choice = input("Enter rock, paper, scissors: ").lower()
    computer_choice = random.choice(choices)

    print("Computer chose:", computer_choice)

    if user_choice == computer_choice:
        print("It's a tie!")

    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        print("You win this round!")
        count_player += 1

    else:
        print("Computer wins this round!")
        count_comp += 1

    print(f"Score -> You: {count_player} | Computer: {count_comp}")
    print("-" * 30)

if count_player == 3:
    print("🎉 Congratulations! You are the overall winner!")
else:
    print("💻 Computer is the overall winner! Better luck next time!")
```
