```python
#1) Write a python program that takes in a student name, class. It should also take in five subject marks of the students and find the total mark and percentage. Display a result in such a way that their name, class,  and percentage are printed.


name = input("Enter student's name: ")
student_class = input("Enter student's class: ")

print("Enter marks for 5 subjects (out of 100):")
sub1 = float(input("Subject 1: "))
sub2 = float(input("Subject 2: "))
sub3 = float(input("Subject 3: "))
sub4 = float(input("Subject 4: "))
sub5 = float(input("Subject 5: "))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total_marks / 500) * 100 

print("\n--- Result ---")
print(f"Name: {name}")
print(f"Class: {student_class}")
print(f"Percentage: {percentage:.2f}%")
```

```python
#2) Input 2 strings concatinate them and store in another variable. then perform generally used string methods on it like

# lower(), upper(), title(), swapcase(), capitalize(), casefold(), center(), count(), endswith(), find(), isalnum(), isdigit(), isnumeric(), isspace(), replace()



str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Concatenate and store in a new variable
combined_str = str1 + " " + str2
print(f"\nConcatenated String: '{combined_str}'\n")

# String Methods
print(f"lower(): {combined_str.lower()}")
print(f"upper(): {combined_str.upper()}")
print(f"title(): {combined_str.title()}")
print(f"swapcase(): {combined_str.swapcase()}")
print(f"capitalize(): {combined_str.capitalize()}")
print(f"casefold(): {combined_str.casefold()}") # Similar to lower(), but more aggressive for some languages
print(f"center(30, '-'): {combined_str.center(30, '-')}")
print(f"count('a'): {combined_str.count('a')}") # Counts occurrences of 'a'
print(f"endswith('s'): {combined_str.endswith('s')}")
print(f"find('a'): {combined_str.find('a')}") # Returns lowest index of 'a', or -1 if not found
print(f"isalnum(): {combined_str.isalnum()}") # False if there's a space or special character
print(f"isdigit(): {combined_str.isdigit()}")
print(f"isnumeric(): {combined_str.isnumeric()}")
print(f"isspace(): {combined_str.isspace()}")
print(f"replace('a', '@'): {combined_str.replace('a', '@')}")
```

```python
#4) Do practise of assignment operators


# Initialize variable
x = 10  
print(f"Initial value (x = 10) -> {x}")

x += 5  # Add AND assign (equivalent to x = x + 5)
print(f"x += 5 -> {x}")

x -= 3  # Subtract AND assign
print(f"x -= 3 -> {x}")

x *= 2  # Multiply AND assign
print(f"x *= 2 -> {x}")

x /= 4  # Divide AND assign (results in a float)
print(f"x /= 4 -> {x}")

x //= 2 # Floor divide AND assign (rounds down to nearest whole number)
print(f"x //= 2 -> {x}")

x %= 2  # Modulus AND assign (remainder of division)
print(f"x %= 2 -> {x}")

# Resetting x for exponentiation
x = 3
x **= 3 # Exponent AND assign (3 to the power of 3)
print(f"x **= 3 (where x was reset to 3) -> {x}")
```

```python
#5) In your last program where you find the total and percentage of a student's marks of 5 subject, find the grade of the student using conditional statement. Eg. grade 'A' if percentage is greator than or equals to 60, 'B' for  percentage is greator than or equals to 50 and less than 60,  'C' for  percentage is greator than or equals to 40 and less than 50,  'D' for  percentage is greator than or equals to 33 and less than 40, otherwise 'Fail'



name = input("Enter student's name: ")
student_class = input("Enter student's class: ")

print("Enter marks for 5 subjects (out of 100):")
sub1 = float(input("Subject 1: "))
sub2 = float(input("Subject 2: "))
sub3 = float(input("Subject 3: "))
sub4 = float(input("Subject 4: "))
sub5 = float(input("Subject 5: "))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total_marks / 500) * 100

# Grading logic
if percentage >= 60:
    grade = 'A'
elif percentage >= 50: 
    grade = 'B'
elif percentage >= 40:
    grade = 'C'
elif percentage >= 33:
    grade = 'D'
else:
    grade = 'Fail'

print("\n--- Final Report Card ---")
print(f"Name: {name}")
print(f"Class: {student_class}")
print(f"Total Marks: {total_marks}/500")
print(f"Percentage: {percentage:.2f}%")
print(f"Grade: {grade}")
```
